```python
ratings_with_books= ratings.merge(books, on="ISBN")
```

```python
num_rating = ratings_with_books.groupby('Book-Title')['Rating'].count().reset_index()
```

```python
final_rating = ratings_with_books.merge(num_rating, on='Book-Title')
```

```python
final_rating = final_rating[final_rating['Rating_y'] >= 50]
```

```python
final_rating.sample(4)
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['Book-Title', 'User_Id'], inplace=True)
```

```python
final_rating.head(1)
```

```python
# Lets create a pivot table
book_pivot = final_rating.pivot_table(columns='User_Id', index='Book-Title', values= 'Rating_x')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors

# Initialize the NearestNeighbors model
model = NearestNeighbors(algorithm='brute')

# Fit the model to your data
model.fit(book_sparse)
```

```python
# Now, let's find the nearest neighbors for a particular row in your pivot table
# Let's say, we want to find the nearest neighbors for the row at index 237
distance, suggestion = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors=6)
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)

    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```
